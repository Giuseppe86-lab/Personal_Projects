First Approach using pandas and numpy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('input_data_lab/data_hooke.csv')
print(df)
```

```python
X = df['L(m)'].values - df['L0(m)'].values
print(X)
y = df['Massa(kg)'].values*9.8
print(y)
```

```python
X_with_intercept = np.column_stack((np.ones(X.shape[0]),X))
print(X_with_intercept)
```

```python
beta = np.linalg.inv(X_with_intercept.T@X_with_intercept)@X_with_intercept.T@y
print(beta)
```

```python
intercept, slope = beta
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
```

```python
y_pred = intercept + slope*X
```

```python
plt.scatter(X, y, label='Dati')
plt.plot(X, y_pred, label = 'Linea di regressione', color = 'red')

plt.title('Regressione Lineare Semplice')
plt.xlabel('Allungamento (m)')
plt.ylabel('Forza Elastica (N)')
plt.legend()
plt.grid(True)
plt.show()
```

Second Approach using scikit-learn

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('input_data_lab/data_hooke.csv')
X = df['L(m)'].values - df['L0(m)'].values
y = df['Massa(kg)'].values*9.8
K = y/X
print(K.mean())
print(K.std())
#Per usare scikit-learn ho bisogno di un vettore colonna per la feature
X=X.reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(X,y)

intercept = model.intercept_
slope = model.coef_[0]
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, label='Dati')
plt.plot(X, y_pred, label = 'Linea di regressione', color = 'red')

plt.title('Regressione Lineare Semplice - con Scikit-Learn')
plt.xlabel('Allungamento (m)')
plt.ylabel('Forza Elastica (N)')
plt.legend()
plt.grid(True)
plt.show()
```
